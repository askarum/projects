```python
from LinearRegression import LinearRegression as linear_reg
import statsmodels.api as sm
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
N_sample = 20
random.seed(101)
X = np.linspace(-5, 5, num = N_sample)
z = np.random.normal(0, 1.2, N_sample)
y = 3*X + 2 + z
```

```python
plt.plot(X, y, 'o')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Synthetic data', pad = 15)
plt.show()
```

```python
model = linear_reg(lr = 0.01, n_iter = 100)
```

```python
model.fit(X.reshape(-1,1), y.reshape(-1,1))
```

```python
y_pred = model.predict(X.reshape(-1,1))
```

```python
plt.plot(X, y_pred.ravel(), color = 'r')
plt.plot(X, y, 'o')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear regression', pad = 15)
plt.show()
```
